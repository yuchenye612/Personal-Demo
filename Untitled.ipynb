# Vehicle Part

```python
# load Data
roads = gpd.read_file('road.zip')
road_nodes = pd.read_csv('CSCL_PUB_Centerline.csv')
```

```python
# Data cleaning: covert the edges into nodes
def coord_lister(geom):
    coords = list(geom.coords)
    return (coords)

coordinates_list = roads.geometry.apply(coord_lister)
roads['u'] = coordinates_list.apply(lambda x: x[0])
roads['v'] = coordinates_list.apply(lambda x: x[1])
```

```python
# small function to check if there is path between 2 points
def find_path(start,end):
    try:
        p = nx.shortest_path(G,source = start, target = end)
        path = nx.path_graph(p, create_using=G) 
        return True
    except:
        return False
```

```python
# The overall vechicle path prediction function 
def find_driving(start,end):
    
        # geting all the nodes and the position of the nodes
    position = roads
    another = roads
    another['positions'] = roads['v']
    another['index'] = roads['v']
    another = another[['index','positions']]
    position['positions'] = roads['u']
    position['index'] = roads['u']
    position = position[['index','positions']]

    position_table = position.append(another, ignore_index=True).set_index('index')
    positions = position_table.to_dict()

    # Build up initial Graph
    G = nx.Graph()

    # get the edges and set the edge weight as the length of edges
    edges_weights = roads[['u','v','shape_leng']]
    edges_weights['edges']= edges_weights[['u','v']].apply(lambda i: (i['u'],i['v']) , axis =1)
    edges_weights['weights']= edges_weights[['shape_leng']].apply(lambda i: {'weight' : i['shape_leng']}, axis =1)
    edges_weights_dict = edges_weights[['edges','weights']].set_index('edges').to_dict()


    G.add_edges_from(roads[['u','v']].apply(lambda i: (i['u'],i['v']) , axis =1))
    nx.set_edge_attributes(G,  edges_weights_dict['weights'])

    other= [i for i in G.edges() ]


    # Build up another graph for plotting

    G_a = nx.Graph()

    edges_weights = roads[['u','v','shape_leng']]
    edges_weights['edges']= edges_weights[['u','v']].apply(lambda i: (i['u'],i['v']) , axis =1)
    edges_weights['weights']= edges_weights[['shape_leng']].apply(lambda i: {'weight' : i['shape_leng']}, axis =1)
    edges_weights_dict = edges_weights[['edges','weights']].set_index('edges').to_dict()


    G_a.add_edges_from(roads[['u','v']].apply(lambda i: (i['u'],i['v']) , axis =1))
    nx.set_edge_attributes(G_a,  edges_weights_dict['weights'])

    
    # We need to consider the situtaion that the starting points or ending points are not in our graph
    # This can be caused by people need to park their car in the parking lots or at their homes
    # So based on these situations we decide to find the 10 closest points from our starting/ending points
    # Then we try to find path between them
    
    # Find 10 closest points to starting/ending points
    if start not in positions['positions'].keys():
        ex = start
        distance2 = position_table
        distance2['x'] = abs(abs(distance2['positions'].apply(lambda x: x[0])) - abs(ex[0]))
        distance2['y'] = abs(abs(distance2['positions'].apply(lambda x: x[1]))- abs(ex[1]))
        distance2['z'] = distance2['x'] + distance2['y']
        z = list(distance2['z'])
        n = list(distance2['positions'])
        nearest_5_start = distance2.nsmallest(n=10, columns=['z'])
    if end not in positions['positions'].keys():
        ex = end
        distance2 = position_table
        distance2['x'] = abs(abs(distance2['positions'].apply(lambda x: x[0])) - abs(ex[0]))
        distance2['y'] = abs(abs(distance2['positions'].apply(lambda x: x[1]))- abs(ex[1]))
        distance2['z'] = distance2['x'] + distance2['y']
        z = list(distance2['z'])
        n = list(distance2['positions'])
        nearest_5_end = distance2.nsmallest(n=10, columns=['z'])
        
    # COnsider the 4 situtaions: Starting/Ending points; In/Not in the Graph
    if start not in positions['positions'].keys():
        if end not in positions['positions'].keys():
            for i in np.arange(10):
                for j in np.arange(10):
                    # Compare the nearest node that to see if there is a possible way
                    if find_path(nearest_5_start.index[i], nearest_5_end.index[j]):
                        p = nx.shortest_path(G,source = nearest_5_start.index[i], target = nearest_5_end.index[j])
                        path = nx.path_graph(p, create_using=G)  
                        break
                    # If the close roads has no possible way we do not have the path and just print to told people the result
                    if i == 9 and j == 9 and find_path(nearest_5_start.index[i], nearest_5_end.index[j]) == False:
                        print('There is no path between these two points.')
                        return
        else:
            for i in np.arange(10):
                # Compare the nearest node that to see if there is a possible way
                if find_path(nearest_5_start.index[i], end):
                    p = nx.shortest_path(G,source = nearest_5_start.index[i], target = end)
                    path = nx.path_graph(p, create_using=G)  
                    break
                # If the close roads has no possible way we do not have the path and just print to told people the result
                if i == 9 and find_path(nearest_5_start.index[i], end) == False:
                    print('There is no path between these two points.')
                    return
        
    else: 
        if end not in positions['positions'].keys():
            for i in np.arange(10):
                # Compare the nearest node that to see if there is a possible way
                if find_path(start, nearest_5_end.index[i]):
                    p = nx.shortest_path(G,source = start, target = nearest_5_end.index[i])
                    path = nx.path_graph(p, create_using=G)  
                    break
                # If the close roads has no possible way we do not have the path and just print to told people the result
                if i == 9 and find_path(start, nearest_5_end.index[i]) == False:
                    print('There is no path between these two points.')
                    return
        else:
            try:
                p = nx.shortest_path(G,source = start, target = end)
                path = nx.path_graph(p, create_using=G)   
            except:
                # If the close roads has no possible way we do not have the path and just print to told people the result
                print('There is no path between these two points.')
                return
            
        
    
    # Calcluate some relevant information   
    all_weight = 0
    for i in list(path.edges):
        if i in edges_weights_dict['weights'].keys():
            all_weight += edges_weights_dict['weights'][i]['weight']
    
    # Time
    time = all_weight/640
    
    distance = all_weight/3200
    
    # Oil consumption
    oil = distance / 22
    
    # Taxi fee
    taxi = 0
    meters = distance * 1600
    if meters < 320:
        taxi += 2.5
    else:
        taxi += 2.5
        taxi += (meters // 320 ) * 0.5 
    print('The time it take should be '+ str(time) + ' min.')
    print('The oil it take should be '+ str(oil) + ' gallon.')
    print('The price if you choose taxi is '+ str(time) + ' $.')
    print('Don\'t forget to calculate the tips.')
    # print out path
    print('The path is '+ str(p))
    
    # Print out graphs
    print('Here is the best way shown on the New York Map')
    plt.figsize=(20, 20)
    nx.draw(G_a, positions['positions'],  node_size=5, node_color ='green',edge_color = 'green')
    nx.draw(path, positions['positions'],node_color='black', node_size= 1, width=10,edge_color = 'red')
    plt.savefig('shortest_path_on_map',figsize=(20, 20))
    plt.show()
    print('Here is the best way only.')
    plt.figsize=(20, 20)
    nx.draw(path, positions['positions'],node_color='red', node_size= 10, width=4,edge_color = 'black')
    plt.savefig('shortest_path',figsize=(20, 20))
    plt.show()


             
```

# Subway Part

```python
def find_subway(entry,ex):
    # Import Data 

    #First Up, we need to use two different datasets of NYC subway for complete network
    #construction. The first dataset loaded as "lnyc" includes the location information
    #of stations, the later "df" embeds the connectivity of stations. The two datasets
    #use differenet names for stations, so first need to construct a mapping tool,
    #so we can connect them into to have enough location information to work with.

    lnyc = gpd.read_file('Subway Stations/geo_export_db260c1f-cfe1-4566-8193-29fc820f2bc3.shp')
    

    # Downloading the csv file from your GitHub

    url = "https://raw.githubusercontent.com/sswarnak77/Shortest-time-between-two-stations-in-New-York-subway-system-using-Python-and-Networkx-/master/New%20York%20subway/final_subway_data.csv"
    download = requests.get(url).content


    df = pd.read_csv(io.StringIO(download.decode('utf-8')))

    listnyc = list(lnyc['name'])

    df.loc[len(df.index)] = ['origin', 'dest', 'dist']
    df.columns = df.iloc[1531]
    g = nx.from_pandas_edgelist(df,source='origin',target='dest',edge_attr='dist')


    a = list(g.edges(data=True))
    b = list(g.nodes())
    c = []
    for i in b:
        c.append(i[2:])

    # Data Cleaning

    c = [elem.replace(' Avenue', ' Ave') for elem in c]
    c = [elem.replace(' Second ', ' 2nd ') for elem in c]
    c = [elem.replace(' Third ', ' 3rd ') for elem in c]
    c = [elem.replace(' Fourth ', ' 4th ') for elem in c]
    c = [elem.replace(' Fifth ', ' 5th ') for elem in c]
    c = [elem.replace(' Sixth ', ' 6th ') for elem in c]
    c = [elem.replace(' Seventh ', ' 7th ') for elem in c]
    c = [elem.replace(' Eighth ', ' 8th ') for elem in c]
    c = [elem.replace(' Ninth ', ' 9th ') for elem in c]
    c = [elem.replace(' Place', ' Pl') for elem in c]
    c = [elem.replace('-', ' ') for elem in c]
    c = [elem.replace('', ' ') for elem in c]
    c = [elem.replace('Avenue ', 'Ave ') for elem in c]
    c = [elem.replace(' Street', ' St') for elem in c]
    c = [elem.replace(' Terminal', ' Term') for elem in c]
    c = [elem.replace(' Ave', ' Av') for elem in c]
    c = [elem.replace(' JFK Airport', '') for elem in c]
    c = [elem.replace(' Boulevard', ' Blvd') for elem in c]
    c = [elem.replace(' Road', ' Rd') for elem in c]
    c = [elem.replace('Roosevelt Island','Roosevelt Island Main St') for elem in c]


    listnyc = [elem.replace(' - ', ' ') for elem in listnyc]
    listnyc = [elem.replace('  ', ' ') for elem in listnyc]
    listnyc = [elem.replace('-', ' ') for elem in listnyc]
    listnyc = [elem.replace('', ' ') for elem in listnyc]
    listnyc = [elem.replace('Ctr', 'Center') for elem in listnyc]
    listnyc = [elem.replace(' Ave', ' Av') for elem in listnyc]

    lnyc['name'] = listnyc
    mapping = dict(zip(g, c))



    #Using input start and target location in coordinates, we use the 'lnyc' dataset
    #to find the stations that are closest to start and target location. We then use
    #the constructed "mapping" variable to communicate to "df" dataset which uses
    #different names for stations. When the equivalent start and target stations are 
    #located in df, a minimum cost shortest path algorithm is used.

    g = nx.relabel_nodes(g, mapping,copy=True)
    edge = list(g.edges(data=True))
    node = list(g.nodes(data=True))

    loc_dict = dict()
    lnyc["x"] = lnyc.centroid.y
    lnyc["y"] = lnyc.centroid.x
    lnyc = lnyc[['x','y','name']]
    lnyc0 = lnyc

    loc_dict = lnyc.set_index('name').to_dict()
    x = loc_dict['x']
    y = loc_dict['y']
    nx.set_node_attributes(g, 0,'x')
    nx.set_node_attributes(g, x,'x')
    nx.set_node_attributes(g, 0,'y')
    nx.set_node_attributes(g, y,'y')




    distance1 = lnyc
    distance0= lnyc
    distance1['x'] = abs(abs(distance1['x']) - abs(entry[0]))
    distance1['y'] = abs(abs(distance1['y'])- abs(entry[1]))
    distance1['z'] = distance1['x'] + distance1['y']

    z = list(distance1['z'])
    n = list(distance1['name'])
    minpos = z.index(min(z))
    mindist1= min(z)
    on_station = n[minpos]

    on_sta = (list(mapping.keys())[list(mapping.values()).index(on_station)])


    lnyc = gpd.read_file('Subway Stations/geo_export_db260c1f-cfe1-4566-8193-29fc820f2bc3.shp')
    lnyc['name'] = listnyc
    lnyc["x"] = lnyc.centroid.y
    lnyc["y"] = lnyc.centroid.x
    lnyc = lnyc[['x','y','name']]

    distance2 = lnyc
    distance2['x'] = abs(abs(distance2['x']) - abs(ex[0]))
    distance2['y'] = abs(abs(distance2['y'])- abs(ex[1]))
    distance2['z'] = distance2['x'] + distance2['y']

    z = list(distance2['z'])
    n = list(distance2['name'])
    minpos = z.index(min(z))
    mindist2= min(z)
    off_station = n[minpos]

    off_sta = (list(mapping.keys())[list(mapping.values()).index(off_station)])

    #From input start and target locations, we return "on_sta" and "off_sta"
    #which are understandable strings by adjacency network graph.

    g = nx.from_pandas_edgelist(df,source='origin',target='dest')


    #Here we modify the attributes of the network. "df" dataset does encode
    #the physical distance between stations by distance attribute, but we wish to
    #account for the inconvinient, bumping experience of transffering subway lines.
    #Using the "diction" parameter, we examine if two connected stations belong to
    #the same subway line. If not, they are connected via transfer, and we increase
    #their distance attribute significantly, so they are much further than they
    #actually are, to account for the difficulty of transferring, and force our
    #program to pick a route involving less transfer if possible.

    a = list(g.edges(data=True))
    b = list(g.nodes())
    a = a[:-1]
    diction = {}

    transit = []

    for i in a:
        l1 = i[0][0]
        l2 = i[1][0]
        if l1 == l2:
            transit.append(0)
        else:
            transit.append(1)

    alpha = 0

    k = df['dist']
    for i in a:

        k = k
        if transit[alpha] == 0:
            diction[tuple([i[0],i[1]])] = float(k[alpha])
        if transit[alpha] == 1:
            diction[tuple([i[0],i[1]])] = float(k[alpha]) + 10000000
        alpha += 1


    nx.set_edge_attributes(g,diction,name='weight')
    #nx.set_edge_attributes(g,1,name='capacity')

    c = list(g.edges(data=True))
    d = list(g.nodes())

    #After modifying edges attribute to account for difficulties in transferring,
    #we apply shortest path to get the list of stations travelled. Using the list,
    #we assume if a passenger stays on a line, it will take 1.5 minutes to go through
    #each station, and it will take 10 minutes to transfer from one line to another,
    #total travelling time is then calculated adding walking times from subway
    #stations to start and destinations. 

    path = nx.all_shortest_paths(g,
                        source='{}'.format(on_sta),
                        target='{}'.format(off_sta),
                        weight = 'weight'
                        )
    path = list(path)[0]
    pathct = []

    for i in path:
        pathct.append(i[0])

    time = len(pathct) * 1.5

    l1 = []

    # taking an counter
    count = 0

    # traversing the array
    for item in pathct:
        if item not in l1:
            count += 1
            l1.append(item)

    time += count * 7
    walkdist = (mindist1 + mindist2) * 900
    time += walkdist

    onlywalk = abs(entry[0]-ex[0])*900 + abs(entry[1]-ex[1])*900


    if onlywalk < time:
        print('Only walking to the destination have better time efficient.')
        print('Time is',onlywalk, "minutes")

    print("For the subway path:")

    print(path)
    print('Time is',time, "minutes")
```
